```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 計算範囲
x = np.linspace(-5, 5, 101)

# 解析解
y_exact = np.exp(x)

# 差分法で求めた近似解
h = 0.01
y_approx = (np.exp(x + h) - np.exp(x)) / h

# グラフの描画
plt.plot(x, y_exact, c='yellow', linewidth=4, label="y = e^x")
plt.plot(x, y_approx, c='blue', linewidth=1, label='difference method')
plt.title('Exponential')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.arange(-5, 6, 1))
plt.yticks(np.arange(-10, 161, 10))
plt.grid(True)
plt.legend()
plt.show()
```
